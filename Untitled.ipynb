```python
import matplotlib
import matplotlib.pyplot as plt
import requests
import json


base_uri = 'https://api.football-data.org/v4/competitions/CL/matches'
headers = { 'X-Auth-Token': '2479cd2aad494af7a80a601e0b6441dc', 'Accept-Encoding': '' }
response = requests.get(base_uri, headers=headers)
matches = response.json()['matches']
matches
```

```python
m_df = pd.DataFrame(matches)
match_df = pd.DataFrame()
match_df['date'] = m_df['utcDate']
match_df['status'] = m_df['status']
match_df['matchday'] = m_df['matchday']
match_df['stage'] = m_df['stage']
match_df['group'] = m_df['group']
match_df['group'] = m_df['homeTeam'].apply(lambda x: x.get('name'))
m_df
```
